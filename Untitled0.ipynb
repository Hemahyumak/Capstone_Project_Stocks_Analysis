Getting SEC data and saving the list of companies to a csv file

```python
import requests
import json
import pandas as pd
from google.colab import files

# Proper SEC API URL for insider trading data
url = "https://www.sec.gov/files/company_tickers.json"

# Add headers to avoid being blocked
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "Accept": "application/json"
}

try:
    response = requests.get(url, headers=headers, timeout=10)

    # Check if the response is valid
    if response.status_code == 200:
        data = response.json()  # Convert to JSON

        # Convert JSON data to DataFrame
        df = pd.DataFrame.from_dict(data, orient="index")

        # Rename columns for clarity
        df.columns = ["CIK", "Ticker", "Company Name"]

        # Save to CSV file
        csv_filename = "sec_company_tickers.csv"
        df.to_csv(csv_filename, index=False)
        print(f"File saved as {csv_filename}")

        # Download the file in Colab
        files.download(csv_filename)

    else:
        print(f"Error: {response.status_code}, Response: {response.text}")

except requests.exceptions.RequestException as e:
    print(f"Network error: {e}")
```

Accessing the list of SEC companies approx 50 for analysis and getting their real time stock data in Yahoo Finance

```python
import pandas as pd
import yfinance as yf
from google.colab import files

# Manually upload the SEC company tickers CSV
uploaded = files.upload()

# Get the uploaded file name dynamically
csv_filename = list(uploaded.keys())[0]

# Load the SEC company tickers from the uploaded CSV
df_sec = pd.read_csv(csv_filename)

# Ensure 'Ticker' column exists before proceeding
if "Ticker" not in df_sec.columns:
    raise ValueError("Error: The uploaded CSV does not contain a 'Ticker' column.")

# Take only the 'Ticker' column (some tickers might not be available in Yahoo Finance)
tickers = df_sec["Ticker"].dropna().unique().tolist()

# Limit the number of tickers to avoid rate limits (example: first 50 tickers)
tickers = tickers[:50]  # Adjust as needed

# Download stock data for the last 6 months with Open, High, Low, Close, Volume
stock_data = yf.download(tickers, start="2023-09-01", end="2024-03-01", group_by="ticker")

# Print sample data to check columns before saving
print("Sample stock data preview:")
print(stock_data.head())

# Validate if essential columns are present before saving
required_columns = ['Open', 'High', 'Low', 'Close', 'Volume']
missing_columns = [col for col in required_columns if col not in stock_data.columns]

if missing_columns:
    print(f"Warning: Missing columns in the downloaded data: {missing_columns}")
else:
    print("All required columns are present.")

# Save stock data to CSV
stock_csv_filename = "yahoo_stock_data_full.csv"
stock_data.to_csv(stock_csv_filename)
print(f"Stock data saved as {stock_csv_filename}")

# Download the file in Google Colab
files.download(stock_csv_filename)
```

Preprocessing

```python
import pandas as pd
from google.colab import files

# 📂 Upload the raw stock data
uploaded = files.upload()

# 📝 Load the file
csv_filename = list(uploaded.keys())[0]
df = pd.read_csv(csv_filename, header=[0, 1], index_col=0, parse_dates=True)

# 🔄 Preprocessing: Handling duplicates and missing values
print("🔍 Missing values before processing:")
print(df.isnull().sum().sum())  # Total missing values

df = df.drop_duplicates()
df.fillna(method='ffill', inplace=True)  # Forward-fill missing values

print("\n✅ Missing values after forward-fill:")
print(df.isnull().sum().sum())  # Check again after filling

# 🏷️ Extract closing prices for all tickers
close_prices = df.xs('Close', axis=1, level=1)

# ---- 1️⃣ Daily Returns Calculation ---- #
daily_returns = close_prices.pct_change().dropna()
daily_returns_csv = "daily_returns.csv"
daily_returns.to_csv(daily_returns_csv)
print(f"✅ Daily Returns saved as {daily_returns_csv}")
files.download(daily_returns_csv)

# ---- 2️⃣ Moving Averages (7-day & 30-day) ---- #
ma_7 = close_prices.rolling(window=7).mean()
ma_30 = close_prices.rolling(window=30).mean()
ma_df = pd.concat([ma_7.add_suffix('_MA7'), ma_30.add_suffix('_MA30')], axis=1).dropna()

ma_csv = "moving_averages.csv"
ma_df.to_csv(ma_csv)
print(f"✅ Moving Averages saved as {ma_csv}")
files.download(ma_csv)

# ---- 3️⃣ Stock Volatility (Standard Deviation of Returns) ---- #
volatility_7 = daily_returns.rolling(window=7).std()
volatility_30 = daily_returns.rolling(window=30).std()
volatility_df = pd.concat([volatility_7.add_suffix('_Volatility7'), volatility_30.add_suffix('_Volatility30')], axis=1).dropna()

volatility_csv = "stock_volatility.csv"
volatility_df.to_csv(volatility_csv)
print(f"✅ Stock Volatility saved as {volatility_csv}")
files.download(volatility_csv)
```

```python
import pandas as pd
from google.colab import files

# 📂 Upload the raw stock data
uploaded = files.upload()

# 📝 Load the file
csv_filename = list(uploaded.keys())[0]
df = pd.read_csv(csv_filename, header=[0, 1], index_col=0, parse_dates=True)

# 🔄 Preprocessing: Handling duplicates and missing values
print("🔍 Missing values before processing:", df.isnull().sum().sum())

df = df.drop_duplicates()
df.ffill(inplace=True)  # Forward-fill missing values

print("\n✅ Missing values after forward-fill:", df.isnull().sum().sum())

# 🏷️ Extract closing prices for all tickers
close_prices = df.xs('Close', axis=1, level=1)

# ---- 1️⃣ Relative Strength Index (RSI) Calculation ---- #
def compute_rsi(data, window=14):
    delta = data.diff(1)
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()

    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

rsi_df = close_prices.apply(compute_rsi)

rsi_csv = "relative_strength_index.csv"
rsi_df.to_csv(rsi_csv)
print(f"✅ Relative Strength Index (RSI) saved as {rsi_csv}")
files.download(rsi_csv)

# ---- 2️⃣ Exponential Moving Average (EMA) Calculation ---- #
ema_7 = close_prices.ewm(span=7, adjust=False).mean()
ema_30 = close_prices.ewm(span=30, adjust=False).mean()
ema_df = pd.concat([ema_7.add_suffix('_EMA7'), ema_30.add_suffix('_EMA30')], axis=1)

ema_csv = "exponential_moving_average.csv"
ema_df.to_csv(ema_csv)
print(f"✅ Exponential Moving Average (EMA) saved as {ema_csv}")
files.download(ema_csv)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# 📂 Upload the stock data CSV file
uploaded = files.upload()

# 📝 Load the file
csv_filename = list(uploaded.keys())[0]
df = pd.read_csv(csv_filename, header=[0, 1], index_col=0, parse_dates=True)

# 🏷️ Extract necessary data
close_prices = df.xs('Close', axis=1, level=1)
volumes = df.xs('Volume', axis=1, level=1)

# ---- 1️⃣ Plot Stock Price Trends ---- #
plt.figure(figsize=(12, 6))
for ticker in close_prices.columns[:5]:  # Plot first 5 tickers for clarity
    plt.plot(close_prices.index, close_prices[ticker], label=ticker)

plt.title("Stock Price Trends Over Time")
plt.xlabel("Date")
plt.ylabel("Closing Price ($)")
plt.legend()
plt.grid(True)
plt.show()

# ---- 2️⃣ Compare Volatility (Rolling Standard Deviation) ---- #
volatility = close_prices.pct_change().rolling(window=14).std()  # 14-day volatility

plt.figure(figsize=(12, 6))
for ticker in volatility.columns[:5]:
    plt.plot(volatility.index, volatility[ticker], label=ticker)

plt.title("Stock Price Volatility (14-day Rolling Std)")
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.legend()
plt.grid(True)
plt.show()

# ---- 3️⃣ Correlation Matrix ---- #
corr_matrix = close_prices.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=False, cmap="coolwarm", linewidths=0.5)
plt.title("Stock Price Correlation Matrix")
plt.show()

# ---- 4️⃣ Volume Analysis (Detect Unusual Spikes) ---- #
volume_spikes = volumes.pct_change().rolling(window=5).max()  # Detect spikes in trading volume

plt.figure(figsize=(12, 6))
for ticker in volume_spikes.columns[:5]:
    plt.plot(volume_spikes.index, volume_spikes[ticker], label=ticker)

plt.title("Unusual Trading Volume Spikes (5-day Rolling Max Change)")
plt.xlabel("Date")
plt.ylabel("Volume Change (%)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from google.colab import files

# 📂 Upload stock data CSV
uploaded = files.upload()
csv_filename = list(uploaded.keys())[0]
df = pd.read_csv(csv_filename, header=[0, 1], index_col=0, parse_dates=True)

# 🏷️ Extract closing prices for a single stock (Example: 'AAPL')
ticker = 'AAPL'
close_prices = df.xs(('Close', ticker), axis=1, level=1).dropna()

# 🔄 Create target variable (shifted price)
close_prices_df = pd.DataFrame(close_prices)
close_prices_df['Target'] = close_prices_df[ticker].shift(-1)  # Predict next day's price

# 🏗️ Feature Engineering: Add Moving Averages
close_prices_df['MA7'] = close_prices_df[ticker].rolling(window=7).mean()
close_prices_df['MA30'] = close_prices_df[ticker].rolling(window=30).mean()

# 🔍 Remove NaN values from moving averages
close_prices_df.dropna(inplace=True)

# 🚀 Splitting Data
X = close_prices_df[[ticker, 'MA7', 'MA30']]
y = close_prices_df['Target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# ---- 1️⃣ Linear Regression ---- #
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

mse_lr = mean_squared_error(y_test, y_pred_lr)
print(f"✅ Linear Regression MSE: {mse_lr:.4f}")

# ---- 2️⃣ Random Forest Regression ---- #
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

mse_rf = mean_squared_error(y_test, y_pred_rf)
print(f"✅ Random Forest MSE: {mse_rf:.4f}")

# ---- 3️⃣ LSTM Neural Network ---- #
X_lstm = np.array(X).reshape(X.shape[0], X.shape[1], 1)
X_train_lstm, X_test_lstm, y_train_lstm, y_test_lstm = train_test_split(X_lstm, y, test_size=0.2, shuffle=False)

model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(X_train_lstm.shape[1], 1)),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train_lstm, y_train_lstm, epochs=10, batch_size=16, verbose=1)

y_pred_lstm = model.predict(X_test_lstm).flatten()
mse_lstm = mean_squared_error(y_test_lstm, y_pred_lstm)
print(f"✅ LSTM MSE: {mse_lstm:.4f}")

# ---- 📊 Visualization: Compare Model Predictions ---- #
plt.figure(figsize=(12, 6))
plt.plot(y_test.index, y_test, label="Actual Prices", color='black')
plt.plot(y_test.index, y_pred_lr, label="Linear Regression", linestyle="dashed")
plt.plot(y_test.index, y_pred_rf, label="Random Forest", linestyle="dotted")
plt.plot(y_test.index, y_pred_lstm, label="LSTM", linestyle="dashdot")
plt.title(f"Stock Price Predictions for {ticker}")
plt.xlabel("Date")
plt.ylabel("Stock Price ($)")
plt.legend()
plt.grid(True)
plt.show()
```
